## 5.1 Epipolar Geometry

### 5.1.1 Property Of Skew-Symmetric Matrix

$$
    [\mathbf{a}]_\times=\left[
        \begin{matrix}
            0 & - a_3 & a_2\\
            a_3 & 0 & -a_1\\
            -a_2 & a_1 & 0 
        \end{matrix}
    \right]
$$

$$
    \mathbf{a}\times \mathbf{b}=
    \left[
        \begin{matrix}
            0 & - a_3 & a_2\\
            a_3 & 0 & -a_1\\
            -a_2 & a_1 & 0 
        \end{matrix}
    \right]
    \left[
        \begin{matrix}
            b_1\\
            b_2\\
            b_3
        \end{matrix}
    \right]=[\mathbf{a}]_\times \mathbf{b}
    =-\mathbf{b}\times \mathbf{a}
    =-[\mathbf{b}]_\times \mathbf{a}
    =[\mathbf{b}]_\times^T\mathbf{a}
    =(\mathbf{a}^T [\mathbf{b}]_\times )^T
$$

$$
\begin{aligned}
    \mathbf{v}^T(\mathbf{Mu}\times\mathbf{w})
    &=\left\lvert\left [ \mathbf{v},\mathbf{Mu},\mathbf{w}\right] \right\lvert
    \\
    &=\left\lvert \mathbf{M} \left [ \mathbf{M}^{-1}\mathbf{v},\mathbf{u},\mathbf{M}^{-1}\mathbf{w}\right] \right\lvert
    \\
    &= \lvert \mathbf{M} \rvert \left\lvert  \left [ \mathbf{M}^{-1}\mathbf{v},\mathbf{u},\mathbf{M}^{-1}\mathbf{w}\right] \right\lvert
    \\
    &= \lvert \mathbf{M} \rvert  (\mathbf{M}^{-1}\mathbf{v})^T(\mathbf{u}\times \mathbf{M}^{-1}\mathbf{w})
    \\
    &= \lvert \mathbf{M} \rvert  \mathbf{v}^T\mathbf{M}^{-T}[\mathbf{u}]_\times\mathbf{M}^{-1}\mathbf{w}
    \\
    &=  \mathbf{v}^T (\lvert \mathbf{M} \rvert \mathbf{M}^{-T}[\mathbf{u}]_\times\mathbf{M}^{-1})\mathbf{w}
    \\
    &=\mathbf{v}^T [\mathbf{Mu}]_\times \mathbf{w}
\end{aligned}
$$

$$
[\mathbf{Mu}]_\times=\lvert \mathbf{M} \rvert \mathbf{M}^{-T}[\mathbf{u}]_\times\mathbf{M}^{-1}
$$

### 5.1.2 Epipolar Geometry

Points $\mathbf{x}$ and $\mathbf{x}'$ are both the images of the 3D point $\mathbf{X}$ on a plane. The set of all such points $\mathbf{x}_i$ in the first image and the corresponding points $\mathbf{x}_i'$ in the second image are projectively equivalent, since they are projectively equivalent to the planar point $\mathbf{x}_i$. Thus there is 2D homography $\mathbf{H}$ mapping $\mathbf{x}_i$ to $\mathbf{x}_i'$.

Given the point $\mathbf{x}'$ the epipolar line $\mathbf{l}'$ passsing through $\mathbf{x}'$ and the epipolar point $\mathbf{e}'$ can be represented by the normal vector of the epipolar plane, since once the plane is determinant, the epiploar line is determinant, i.e. $\mathbf{l}'=\mathbf{e}'\times \mathbf{x}'=[\mathbf{e}']_\times \mathbf{x}'$. i.e. $\mathbf{l}'$ is the normal vector of the epipolar line. Since $\mathbf{x}'$ may be rewritten as $\mathbf{H}\mathbf{x}$, we have:
$$
    \mathbf{l}'=[\mathbf{e}']_\times\mathbf{Hx}=\mathbf{Fx}
$$

where we define $\mathbf{F}=[\mathbf{e}']_\times \mathbf{H}$, the fundamental matrix. Obviously, $\mathbf{H}=\mathbf{P}'\mathbf{P}^+$, and $\mathbf{PP}^+=\mathbf{I}$.

Suppose the camera matrices are those of calibrated stero rig with the world origin at the first camera:
$$
    \mathbf{P}=\mathbf{K}[\mathbf{I} \, \vert \, \mathbf{0}], \mathbf{P}'=\mathbf{K}'[\mathbf{R}\,\vert\,\mathbf{t}]
$$

Then:
$$\mathbf{P}^+=\left[ \begin{matrix} \mathbf{K}^{-1} \\ \mathbf{0}^T \end{matrix} \right], \mathbf{c}=\left[\begin{matrix} \mathbf{0} \\1 \end{matrix} \right],
\mathbf{t}=-\mathbf{Rc}', 
\mathbf{e}=\mathbf{P}\left[ \begin{matrix} \mathbf{c}'\\1 \end{matrix} \right]=-\mathbf{KR}^T\mathbf{t}
$$

Note: Cross product can be seen as a vector rotating about another vector.

$$
\begin{aligned}
    \mathbf{F}
    &=[\mathbf{P}'\mathbf{c}]_\times \mathbf{P}'\mathbf{P}^+
    \\
    &=[\mathbf{K}'\mathbf{t} ]_\times \mathbf{K}'\mathbf{RK}^{-1}
    \\
    &=\alpha' f'^2\mathbf{K}'^{-T}[\mathbf{t}]_\times\mathbf{K}'^{-1}\mathbf{K}'\mathbf{RK}^{-1}
    \\
    &=\alpha' f'^2\mathbf{K}'^{-T}[\mathbf{t}]_\times\mathbf{RK}^{-1}
    \\
    &=\alpha' f'^2\mathbf{K}'^{-T}\mathbf{R}[\mathbf{R}^T\mathbf{t}]_\times\mathbf{K}^{-1}
    \\
    &=\alpha' f'^2\mathbf{K}'^{-T}\mathbf{R}\frac{1}{\alpha f^2}\mathbf{K}^T[\mathbf{KR}^T\mathbf{t}]_\times
    \\
   &=\alpha' f'^2\mathbf{K}'^{-T}\mathbf{R}\frac{1}{\alpha f^2}\mathbf{K}^T[-\mathbf{e}]_\times
\end{aligned}
$$

Ignore the scalar, then:
$$
\mathbf{F}=\mathbf{K}'^{-T}\mathbf{RK}^T[\mathbf{e}]_\times=\mathbf{K}'^{-T}[\mathbf{t}]_\times\mathbf{RK}^{-1}
$$

It is easy to prove that:
$$
    \mathbf{F}\mathbf{e}=\mathbf{0},\mathbf{e}'^T\mathbf{F}=\mathbf{0}
$$

And for the corresponding image points $\mathbf{x},\mathbf{x}'$ of the scene point $\mathbf{X} \in \mathbb{R}^3$: 
$$
\begin{aligned}
    \mathbf{x}'^T\mathbf{Fx}
    &=(\mathbf{K}'\mathbf{RX}+\mathbf{K}'\mathbf{t})^T[\mathbf{t}]_\times \mathbf{RK}^{-1}\mathbf{KX}
    \\
    &=(\mathbf{RX})^T[\mathbf{t}]_\times\mathbf{RX}
    \\
    &=0
\end{aligned}
$$

The epipolar line passing through $\mathbf{e}$ and $\mathbf{x}$ in the first image can be expressed as the parameterized equation:
$$
    t\mathbf{x}+(1-t)\mathbf{e}
$$

$$
    \mathbf{x}'^T\mathbf{F}(t\mathbf{x}+(1-t)\mathbf{e})= t\mathbf{x}'^T\mathbf{F}\mathbf{x}+ (1-t)\mathbf{x}'^T\mathbf{F}\mathbf{e}=0
$$

Solving $\mathbf{x}'^T\mathbf{Fx}=0$ inducing to get the points in the epipolar line in the first image when knowing $\mathbf{x}'$.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
class Camera:
    def __init__(self,P):
        self.P=P    # 投影矩阵
        self.K=None # 标定矩阵
        self.R=None # 旋转矩阵
        self.t=None # 平移
        self.c=None # 照相机中心
        
    def project(self,X):
        """
        Args:
            X (np.ndarray(float)[4,n])
        """
        x=self.P @ X # [3,n]
        x=x/x[2]
        
        return x

    def factor(self):
        """
        P=K[R | t]=[KR | Kt]
        """
        # 1. rq decomposition
        
        # use qr to get rq
        R,K=np.linalg.qr( np.flipud(self.P[:,:3]).T)
        R=np.flipud(R.T)
        K=np.fliplr(np.flipud(K.T))
        
        # use scipy interface
        #K,R = scipy.linalg.rq(self.P[:,:3])
        
        # 2. guarantee the sign of the inner parameter matrix since the result of the qr decomposition is not unique.
        T=np.diag(np.sign(np.diag(K)))
        
        if np.linalg.det(T)>0:
            T[1,1]*=-1
        
        # 3. get k
        K = K @ T
        
        # 4. get R , TT=I
        R =T @ R
        
        # 5. get t
        t = np.linalg.inv(K) @ self.P[:,3]
        
        self.K=K
        self.R=R
        self.t=t
        
        return K,R,t
    
    def center(self):
        if self.c is not None:
            return self.c
        self.c=-self.R.T@self.t
        return self.c
```

### The Eight Point Algorithm

$\mathbf{x}'^T\mathbf{F}\mathbf{x}=0$,$\mathbf{F} \in \mathbb{R}^{3\times 3}$.

$$
\begin{aligned}
    \mathbf{x}'^T\mathbf{F}\mathbf{x}
    &=\mathbf{x}'^T\left[\mathbf{f}_1,\mathbf{f}_2,\mathbf{f}_3\right]\left[ \begin{matrix} x_1\\x_2\\x_3 \end{matrix} \right]
    \\
    &=\mathbf{x}'^T\mathbf{f}_1x_1+\mathbf{x}'^T\mathbf{f}_2x_2+\mathbf{x}'^T\mathbf{f}_3x_3
    \\
    &=\left[ \mathbf{x}'^T x_1 ,\mathbf{x}'^T x_2 , \mathbf{x}'^T x_3  \right] \left[ \begin{matrix}\mathbf{f}_1 \\ \mathbf{f}_2 \\ \mathbf{f}_3 \end{matrix} \right]
\end{aligned}
$$

Since the freedom of $\mathbf{F}$ is $8$, then we need more than eight points to calculate matrix $\mathbf{F}$:

$$
    \left[ 
        \begin{matrix}
            \mathbf{x}^{1'T}x^1_1 &\mathbf{x}^{1'T}x^1_2 &\mathbf{x}^{1'T}x^1_3
            \\
            \mathbf{x}^{2'T}x^2_1 &\mathbf{x}^{2'T}x^2_2 &\mathbf{x}^{2'T}x^2_3
            \\
            \vdots & \vdots & \vdots\\
            \mathbf{x}^{n'T}x^n_1 &\mathbf{x}^{n'T}x^n_2 &\mathbf{x}^{n'T}x^n_3
        \end{matrix}
    \right]
    \left[ \begin{matrix}\mathbf{f}_1 \\ \mathbf{f}_2 \\ \mathbf{f}_3 \end{matrix} \right]
    =\mathbf{Af}=\mathbf{0}
$$

```python
def compute_fundamental(x1,x2):
    """
    Args:
        x1 (np.ndarray(float)[3,n])
        x2 (np.ndarray(float)[3,n])
    Return f
    """
    assert x1.shape==x2.shape
    _,n=x1.shape
    
    # 1. construct A
    x1=x1.T
    x2=x2.T
    A=x1[:,None]*x2[...,None] # [n,3,3]
#     A=A.T
#     A=x1[...,None]*x2[:,None] # [n,3,3]

    A=A.reshape(n,-1)

#     A = np.zeros((n,9))
#     for i in range(n):
#         A[i] = [x1[0,i]*x2[0,i], x1[0,i]*x2[1,i], x1[0,i]*x2[2,i],
#         x1[1,i]*x2[0,i], x1[1,i]*x2[1,i], x1[1,i]*x2[2,i],
#         x1[2,i]*x2[0,i], x1[2,i]*x2[1,i], x1[2,i]*x2[2,i] ]
    
    # 2. get f
    u,s,v=np.linalg.svd(A)
    f=v[-1]
    f=f.reshape(3,3)
    
    # 3. make rank 2
    u,s,v=np.linalg.svd(f)
    s[-1]=0
    f=u@ np.diag(s) @ v
    
    # calcute x'Fx=0, so need transpose
    return f.T
```

```python
def compute_epipole(f):
    u,s,v=np.linalg.svd(f)
    epipole=v[-1]
    epipole=epipole/epipole[2]
    return epipole
```

```python
def plot_epipolar_line(im,F,x,epipole=None,show_epipole=True):
    """ Plot the epipole and epipolar line F*x=0
    in an image. F is the fundamental matrix
    and x a point in the other image."""
    m,n = im.shape[:2]
    line = F@x
    # epipolar line parameter and values
    t = np.linspace(0,n,100)
    lt = (line[0]*t+line[2])/-line[1]
    # take only line points inside the image
    ndx = (lt>=0) & (lt<m)
    plt.plot(t[ndx],lt[ndx],linewidth=2)
    if show_epipole:
        if epipole is None:
            epipole = compute_epipole(F)
        plt.plot(epipole[0]/epipole[2],epipole[1]/epipole[2],'r*')
```

### 5.2.1 Triangulation (3D Reconstruction)

$$
    \left[
    \begin{matrix}
        \mathbf{P}_1 & \mathbf{x}_1 & \mathbf{0}\\
        \mathbf{P}_2 & \mathbf{0} & \mathbf{x}_2
    \end{matrix}
    \right]
    \left[
    \begin{matrix}
        \mathbf{X}\\
        \lambda_1\\
        \lambda_2
    \end{matrix}
    \right]=\mathbf{0}
$$

```python
def triangulate_point(x1,x2,P1,P2):
    """
    Trigulate
    Args:
        x1 (np.ndarray([3]))
        x2 (np.ndarray([3]))
    Return:
        X (np.ndarray([4])): the reconstrcuted 3D one 
    """
    A=np.zeros((6,6))
    A[:3,:4]=P1
    A[3:,:4]=P2
    A[:3,4]=x1
    A[3:,5]=x2
    
    u,s,v=np.linalg.svd(A)
    X=v[-1,:4]
    X=X/X[3]
    
    return X
```

```python
def triangulate(x1,x2,P1,P2):
    """ Two-view triangulation of points in
    x1,x2 (3*n homog. coordinates). 
    """
    n = x1.shape[1]
    if x2.shape[1] != n:
        raise ValueError("Number of points don't match.")
    X = [ triangulate_point(x1[:,i],x2[:,i],P1,P2) for i in range(n)]
    return np.array(X).T
```

### 5.2.2 Computing The Camera Matrix From 3D Points (Camera Resectioning)

$$
    \begin{aligned}
        \mathbf{PX}
        &=\lambda\mathbf{x}
        =\left[\begin{matrix} \mathbf{p}_1\\\mathbf{p}_2\\\mathbf{p}_3 \end{matrix} \right]
        \mathbf{X}
       =\left[\begin{matrix} \mathbf{X}^T\mathbf{p}_1^T\\ \mathbf{X}^T\mathbf{p}_2^T\\ \mathbf{X}^T\mathbf{p}_3^T \end{matrix} \right]
        = \lambda\left[\begin{matrix} x\\ y \\ 1  \end{matrix} \right]
       \\ 
        &\Rightarrow
        \left[\begin{matrix} \mathbf{X}^T\mathbf{p}_1^T -\lambda x \\ \mathbf{X}^T\mathbf{p}_2^T-\lambda y\\ \mathbf{X}^T\mathbf{p}_3^T-\lambda  \end{matrix} \right]
       \\
       &=\left[
           \begin{matrix}
               \mathbf{X}^T & \mathbf{0}^T &\mathbf{0}^T & -x\\
               \mathbf{0}^T & \mathbf{X}^T  & \mathbf{0}^T & -y\\
               \mathbf{0}^T & \mathbf{0}^T  & \mathbf{X}^T & -1\\
           \end{matrix}
       \right] 
       \left[
           \begin{matrix}
               \mathbf{p}_1^T\\
               \mathbf{p}_2^T\\
               \mathbf{p}_3^T\\
               \lambda
           \end{matrix}
       \right]=\mathbf{0}
    \end{aligned}
$$

Generalize to $N$ Points:
$$
 \left[
           \begin{matrix}
               \mathbf{X}_1^T & \mathbf{0}^T &\mathbf{0}^T   & -x_1 & 0 & \cdots\\
               \mathbf{0}^T & \mathbf{X}_1^T  & \mathbf{0}^T & -y_1  & 0 & \cdots\\
               \mathbf{0}^T & \mathbf{0}^T  & \mathbf{X}_1^T & -1   & 0 & \cdots\\
               \mathbf{X}_2^T & \mathbf{0}^T &\mathbf{0}^T   &  0 & -x_2 & \cdots\\
               \mathbf{0}^T & \mathbf{X}_2^T  & \mathbf{0}^T & 0 & -y_2& \cdots\\
               \mathbf{0}^T & \mathbf{0}^T  & \mathbf{X}_2^T & 0 & -1 & \cdots\\
               \vdots     & \vdots     & \vdots       & \vdots  & \vdots &\ddots 
           \end{matrix}
       \right] 
       \left[
           \begin{matrix}
               \mathbf{p}_1^T\\
               \mathbf{p}_2^T\\
               \mathbf{p}_3^T\\
               \lambda_1\\
               \lambda_2\\
               \vdots
           \end{matrix}
       \right]=\mathbf{Mv}=\mathbf{0},
$$

```python
def compute_P(x,X):
    """
    Compute camera matrix `\bm{P}`
    Args:
        x (np.ndarray(float)[3,n]):
        X (np.ndarray(float)[4,n]):
    Return:
        P (np.ndarray(float)[3,4])
    """
    _,n =x.shape
    assert X.shape[1] == n
    M = np.zeros((3*n,12+n))
    for i in range(n):
        M[3*i,0:4] = X[:,i]
        M[3*i+1,4:8] = X[:,i]
        M[3*i+2,8:12] = X[:,i]
        M[3*i:3*i+3,i+12] = -x[:,i]
    U,S,V = np.linalg.svd(M)
    return V[-1,:12].reshape((3,4))
```

#### 5.2.3.1 Uncalibrated Case (Projective Reconstruction)
